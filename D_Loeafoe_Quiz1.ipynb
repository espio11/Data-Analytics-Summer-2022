# **Data Summary**
### **Attribute 0: Carat**
The attribute of column 0 is the carat. A ‘carat’ is a diamond industry specific unit of measurement used to specify the weight of a diamond. One carat is equal to 200 milligrams, for example, a 5-carat stone will weigh 1 gram. The values of the carat column are of the floating point number data type. The carat of the diamonds in the table range from .2 to 5.01 carat. Knowing the carat of a diamond is important because the weight of a precious stone factors into its price. Heavier stones will likely cost more once the other attributes are factored in.

### **Attribute 1: Cut**
The attribute of column 1 is the cut. Simply put, the cut of a diamond is a measure of how well a diamond’s facets interact with light. The cut of a diamond has direct influence on a diamond's other attributes such as brightness, fire, scintillation, brightness, sparkle, etc. The values of the cut column are of the String data type. The values range from 'fair' to 'ideal'. It is important to know the cut of a diamond because it directly influences multiple important attributes that will all be taken into consideration when determining the price.

### **Attribute 2: Color**
The attribute of column 2 is color. The color of a diamon is the subtle coloring the stone takes on due to impurities they pick up in the earth. The color of a diamon is indicative of its rarity and quality. Completely colorless diamonds are the rarest, which makes them valuable. There are also other colors that fall within a particular value range. The values of the color column are of the String data type. The unique values of this column are listed below with a brief description of each.

* D - Absolutely colorless or icy white. The highest color grade—extremely rare and most expensive.
* E - Colorless. Only miniscule traces of color can be detected by an expert gemologist—a rare, high quality diamond.
* F - Colorless. Slight color detected by an expert gemologist, but still considered a “colorless” grade—a high-quality diamond.
* G - Near-colorless. Color noticeable when compared to diamonds of better grades, but offers excellent value.
* H - Near-colorless. Color noticeable when compared to diamonds of better grades, but offers excellent value.
* I - Near-colorless. Slightly detected color—a good value.
* J - Near-colorless. Slightly detected color—a good value.

The color of a diamond is an attribute that is taken into consideration when determining the rarity and quality of a diamond. Rare and higher quality diamonds will always cost more.

### **Attribute 3: Clarity**

The clarity of a diamond refers to the purity and rarity of the stone, and the degree to which it presents blemishes and inclusions. When forming, natural microscopic characteristics can be trapped within or on the diamond. These internal and surface characteristics are inspected by gemologists, who use 10x magnification and a qualitative grading system to assign a numerical value (clarity grade) to each diamond.

The lower the number of imperfections and flaws in a diamond’s aesthetic appearance, the higher the clarity grade. When determining the optimum clarity for a diamond, keep in mind that no diamond is entirely pure. However, the closer it gets to purity, the clearer it becomes. Diamonds with no or few inclusions are considered particularly rare and highly valued. While clarity has a significant effect on the value of a diamond, most defects are invisible to the naked or unassisted eye.

The values of the clarity column are of the String data type. The unique values of this column are listed below with a brief description of each.

* IF - 	Internally Flawless
* VVS1/VVS2 - Very Very Slightly Included
* VS1/VS2 - Very Slightly Included
* SI1/SI2 - Slightly Included
* I1 - Included

it is important to know the clarity of a diamond beacuse the rarity and quality of a diamond is inversely propotional to the number of flaws it has. In other words, the price of a diamond goes up as the number of flaws decreases.

### **Attribute 4: Depth**

The depth of diamond is its height (in millimeters) measured from the culet (bottom tip) to the table (flat, top surface). On a grading report, there are normally two measurements of diamond depth. The first measurement is the actual depth of diamond in millimeters. The second is the diamond depth percentage, indicating the diamond depth in relation to its width. The diamond depth percentage is calculated by dividing total depth by the average diameter, then multiplying by 100. The values of the depth column are of the floating point number data type. The values range from 43.0 to 79.0. 

It is important to know the depth of a diamond because it determines how much of that light reflects back up to the observer’s eye. If the diamond depth is too shallow, the light strikes the pavilion facet at a low angle and passes through the facet, escaping through the bottom of the diamond. If the diamond depth is too deep, the entering light reflects off the first pavilion facet but strikes the second pavilion at too low an angle, causing the light to refract, escaping through the bottom of the diamond. A diamond with an ideal diamond depth will allow the light to strike each pavilion facet at an angle which allows most of the light to reflect back to the top. In short, the depth of a diamond affects how much light is reflected. Of course, a diamond the shines brighter will sell for more.

### **Attribute 5: Table**
A diamond’s table refers to the flat facet of the diamond which can be seen when the stone is face up. As the largest facet on a diamond, a diamond’s table plays a vital role in creating the diamond’s brilliance by refracting entering light rays and allowing reflected light rays back into the observer’s eye. The values of the table colum are of the floating point number data type. The values range from 43.0 to 95.0. Similar to the depth, the table of a diamond has a direct effct on the diamond's brilliance. Diamonds with an ideal table shine brighter and will sell better.

### **Attribute 6: Price**
The price of a diamond is the cost to aquire it or the price at which it should be sold. The price is determined by weighing and considering all of the attributes of a diamond. Typically speaking, diamonds of higher quality and rarity sell for a higher price. It is important to know the price of a diamond so it can be sold for a price that is competitive but also profitable to the selling organization. The values of the price column are of the integer data type. The values range from 326 to 18823. If the true price of a diamond is not known, one runs the risk of underselling the diamond which is essentially an unprofitable transaction. Knowing the price of the diamond solves this problem.

### **Attributes 7,8,9: X,Y,Z**
The x,y and z of a diamond are measurements of the diamonds length, width and physical depth respectively. These numbers are important because the describe how a diamond's mass is physically distributed in relation to its shape. It is important to know these numbers in order to properly appraise the diamond's shape and size. The values of the x, y and z lists are of the floating point number data type. x values range from 0.0 to 10.74. y values range from 0.0 to 58.9. z values range from 0.0 to 31.8.

Unexpected entries found in numerical data will be replaced by the median and unexpected entries found in categorical data will be replaced by the mode. I will also be generating summary statistics for each category in the following cells along with an appropriate graph for each attribute.

```python
#Column 0: Carat

carat_list = df["carat"] #Place column in a list
carat_set = set() #Add unique values to a set
for i in range(len(carat_list)):
  carat_set.add(carat_list[i])

print(sorted(carat_set)) #Sort and print unique values

null_count = carat_list.isnull().sum() #Count the number of null values. Will be handled if the count is not 0.
print("Number of Null values:", null_count) #Print count

nan_count = np.isnan(carat_list).sum() #Count the number of Nan values. Will be handled if the count is not 0.
print("Number of Nan values: ", nan_count) #Print count

print("\nMinimum Value:", min(carat_list)) #Print min
print("Maximum Value:", max(carat_list)) #Print max

#No unexpected entries were found so no replacements were made

print("Mean:", statistics.mean(carat_list)) #Print mean
print("Standard Deviation:", statistics.stdev(carat_list)) #Print standard deviation
print("Median:", statistics.median(carat_list)) #Print median
print("Mode:", statistics.mode(carat_list)) #Print mode
```

# **Exploritory Data Analysis Findings**
### **Carat**
Carat has a positive correlation with the x, y and z attributes. This makes sense since the physical weight(carat) of a diamond increases, so shall its dimensions. Carat also has a strong correlation with the price attribute. This is to be expected since a heavier diamond will likely be sold for a higher price.
### **Depth**
Depth does not seem to form any meaningful relationship with the other attributes.
### **Table**
Table does not seem to form any meaningful relationship with the other attributes.
### **Price**
Price forms a positive correlation with the x attribute. The x attribute is a measure of the height of a gem. Naturally, "taller" gems tend to be of bigger size and will cost more. Price also has a positive correlation with carat. heavier gems tend to be bigger and will cost more.
### **X, Y & Z**
These last 3 attributes tend to behave similarly since they are strongly linked. First off, all 3 attributes form a positive correlation with each other because they all must increase proportionately with the size of the diamond. An increase in 1 will cause an increase in the other 2 attributes. X, Y & Z also have a positive correlation with price and carat. This is to be expected because heavier diamonds will have bigger dimensions of measurement and big diamonds with big dimensions come with a heavier price tag.

```python
# q1) What proportion of diamonds are between .30 and 1.08 carats?

count = 0 #Variable to count how many values satisfy my future conditions

for i in range(len(carat_list)): #Iterate through list
  if carat_list[i] >= .30 and carat_list[i] <= 1.08: #If current value is greater than or equal to .30 and it less than or equal to 1.08
    count += 1 #Increment counter

print(count/len(carat_list)) #Print proportion: number of values that satisfy the conditions / all values
```

```python
# q2) How many of the diamonds have equal x and y dimensions?

count = 0 #Variable to count how many values satisfy my future conditions

for i in range(len(x_list)): #Iterate through list
  if x_list[i] == y_list[i]: #If the current value of both lists are equal
    count += 1 #Increment counter

print(count) #Print count
```

```python
# q3) How many of the diamonds have a carat value that is less than the mean carat value?

carat_mean = statistics.mean(carat_list)
count = 0 #Variable to count how many values satisfy my future condition

for i in range(len(carat_list)): #Iterate through list
  if carat_list[i] < carat_mean: #If current value of carat_list is less than carat_mean
    count += 1 #Increment counter

print(count) #Print count
```

```python
# q4) How many diamonds have a Premium cut or better? 
# Note that the possible values for the quality of a cut are ranked in ascending order as follows: Fair / Good / Very Good / Premium / Ideal

count = 0 #Variable to count how many values satisfy my future condition

for i in range(len(cut_list)): #Iterate through list
  if cut_list[i] == "Premium" or cut_list[i] == "Ideal":
    count += 1 #Increment counter

print(count) #Print count
```

```python
# q5) Which diamond has the highest price per carat? What is its value?

high_number = 0 #Variable to keep track of the number of the diamond
high_ppc = 0 #Variable to keep track of the highest price per carat

for i in range(len(carat_list)): #Iterate through list
  if (price_list[i]/carat_list[i]) > high_ppc: #If the price per carat of the current diamond is higher than the saved highest price per carat
    high_number = i #Store the number of the new diamond with the highest price per carat
    high_ppc = (price_list[i]/carat_list[i]) #Store the value of the new diamond with the highest price per carat

print("The diamond with the highest price per carat is diamond number", high_number, "with a value of", high_ppc)
```
